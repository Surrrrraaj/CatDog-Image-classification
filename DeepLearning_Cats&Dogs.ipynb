```python
image_size = 250
```

```python
input_image = []

for i in cate:
    folders= os.path.join(path1,i)
    label = cate.index(i) #here we need tom tell our model whic is cats&dogs
    print('Valuve for label is ', label)
    for image in os.listdir(folders):
        image_path = os.path.join(folders, image)
        image_array = cv2.imread(image_path) #image converted into array format
        image_array = cv2.resize(image_array, (image_size, image_size)) #resizing each image to 250-250
        input_image.append([image_array, label])   #we are attaching the target variable
        plt.imshow(image_array)
       
```

```python
len(input_image)
```

```python
np.random.shuffle(input_image)
```

```python
input_image
```

```python
#till this point the x & Y values are together lets seperate them.

X = []
Y = []

for X_values, labels in input_image:
    X.append(X_values)
    Y.append(labels)
```

```python
X #we can see only X values here
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
plt.imshow(X[1])
```

```python
from keras.models import Sequential
from keras.layers import Conv2D , MaxPool2D, Flatten, Dense, Dropout
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
model = Sequential()

model.add(Conv2D(filters = 8 , kernel_size=(5,5), input_shape=(250,250,3), activation ='relu', padding='same'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X,Y, epochs=10)
```

```python
model.summary()
```

```python
pred_values = model.predict(X)
```

```python
pred_values #probality values 
```

```python
pred_classes = np.argmax(pred_values, axis=1)
```
